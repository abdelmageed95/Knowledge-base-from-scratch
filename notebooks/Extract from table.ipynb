```python
from pandas.io.html import read_html
import re
from sklearn.utils import shuffle
```

```python
def extract_table(page_url):
  wikitables = read_html(page_url,index_col=1 , match = "Appearances and goals by club, season and competition", attrs={"class":"wikitable"})
  page_name = re.findall(r'/(\w+)', page_url)[-1].replace("_", " ") # extract page name from url
  table = wikitables[0]
  table.columns.values[0] = 'club'
  table.reset_index(inplace = True)
  table.columns.values[0] = 'Season'
  #table.drop( table.columns[11:13] , inplace = True , axis = 1)
  col_names = []

  # clean column names (to be general for that template)
  for i in  table.columns:
    if type(i) is tuple:
      col = ' '.join(i)
      col = ' '.join(re.sub(r"\[.*?\]","", col).strip().split(" "))
      col = ' '.join(re.sub(r"[\([{})\]]|","", col).strip().split(" "))
      col_names.append(col.lower())
    else:
      col_names.append(i.lower())
  table.columns = col_names
  return  page_name ,table
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    def add_edge(self, vertex1, vertex2 ):
        self.graph[vertex1].append(vertex2)
        #self.graph[vertex2].append(vertex1)


    def remove_edge(self, vertex1, vertex2):
        self.graph[vertex1].remove(vertex2)
        #self.graph[vertex2].remove(vertex1)

    def remove_vertex(self, vertex):
        for v in self.graph[vertex]:
            self.graph[v].remove(vertex)
        del self.graph[vertex]

    def get_neighbors(self, vertex):
        return self.graph[vertex]

    def get_vertices(self):
        return self.graph.keys()

    def the_graph(self):
      return self.graph

        
```

```python
def season_detail(summary_table , player_name):
  Triplets = []
  df = summary_table[ summary_table['season'] != 'Total' ].iloc[:-1] 
  for i in range(len(df)):
    cl =  df['club'].iloc[i]
    CLD = [ cl , "part of" , df['league division'].iloc[i] ]
    if CLD not in Triplets: 
      Triplets.append(CLD)
  
  for i in range(len(df)):
      Triplets.append([player_name , "total matches in season {} ".format(df['season'].iloc[i]) , df['total apps'].iloc[i] ])
      Triplets.append([player_name , "total goals in season {} ".format(df['season'].iloc[i]) , df['total goals'].iloc[i] ])

      Triplets.append([player_name , "total matches in league in season {} ".format(df['season'].iloc[i]) ,df['league apps'].iloc[i]  ])
      Triplets.append([player_name , "total goals in league in season {} ".format( df['season'].iloc[i]) , df['league goals'].iloc[i] ])

      Triplets.append([player_name , "total matches in national c in season {} ".format( df['season'].iloc[i]) , df['national cup apps'].iloc[i]  ])
      Triplets.append([player_name , "total goals in national cup in season {} ".format( df['season'].iloc[i]) , df['national cup goals'].iloc[i] ])

      Triplets.append([player_name , "total matches in league cup in season {} ".format( df['season'].iloc[i]) ,df['league cup apps'].iloc[i] ])
      Triplets.append([player_name , "total goals in league cup in season {} ".format(df['season'].iloc[i]) , df['league cup goals'].iloc[i] ])

      Triplets.append([player_name , "total matches in continental competition in season {} ".format(df['season'].iloc[i]) , df['continental apps'].iloc[i] ])
      Triplets.append([player_name , "total goals in continental competition in season {} ".format(df['season'].iloc[i]) , df['continental goals'].iloc[i] ])

  return Triplets
```

```python
def player_timeline(summary_table, page_name):
  Triplets = []
  V2s = [] # vertex of two elements as tuple ( season , clup name)
  for j in range(len(summary_table)): 
    row = [i for i in summary_table.iloc[j]]
    if (row[0] != 'Total') and (row[0] != 'Career total'):
      ver = ','.join([row[0] ,row[1]])
      V2s.append(ver)

  t_graph = Graph()
  t_graph.add_vertex(page_name)
  for i in V2s:
    t_graph.add_vertex(i)
  for J in range(len(V2s)):
    t_graph.add_edge(page_name , V2s[J])

  edges = t_graph.the_graph()
  edge = {page_name : edges[page_name]}
  seasons_clubs = edges[page_name]
  for i in seasons_clubs :
    i = i.split(',')
    T1= [page_name , 'play season '+ i[0] + ' for',i[1]]
    Triplets.append(T1)
  return Triplets
```
